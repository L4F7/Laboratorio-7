```python
# Estudiante: Luis Fuentes Fuentes

import folium
from folium.plugins import MiniMap
import pandas as pd

# Make a data frame with dots to show on the map
data = pd.DataFrame({
   'lat':[9.998819773228847, 9.970888331829805, 10.019755721679521, 10.31908674197246, 8.573345115014337, 9.379729328198778, 10.617496255140978, 10.13458161347997],
   'lon':[-84.11147530919659, -84.12840277422087, -84.19701576339762, -83.92275721662864, -82.87243478473953, -83.69151440207727, -85.45141746126018, -85.4466021083532],
   'sede':['Campus Omar Dengo', 'Campus Benjamín Núñez', 'Sede Interuniversitaria', 'Campus Sarapiquí', 'Campus Coto', 'Campus Pérez Zeledón', 'Campus Liberia', 'Campus Nicoya'],
}, dtype=str)

CR = folium.Map(location=(9.978778639068794, -84.17256532506566), zoom_start = 8)

# add marker one by one on the map
for i in range(0,len(data)):
   folium.Marker(
      location=[data.iloc[i]['lat'], data.iloc[i]['lon']],
      popup=data.iloc[i]['sede'],
      icon=folium.Icon(color="red", prefix='fa',icon='university')
   ).add_to(CR)

minimap = MiniMap()
CR.add_child(minimap)

CR
```

```python
CR.save('C:X.html')
```
