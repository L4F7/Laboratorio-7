```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("casasboston.csv")
```

```python
print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
```

```python
print(df)
```

```python
graph = df.N_HABITACIONES_MEDIO.plot.hist(color='lightGreen', rwidth=0.9, edgecolor = 'black')

graph.set_title("Distribución de promedio de número de habitaciones", size = 16)

graph.set_xlabel("Número de habitaciones", labelpad=20, weight='bold', size=12)

graph.set_ylabel("Frecuencia", labelpad=20, weight='bold', size=12)

plt.show()
```

```python
scatter = df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN", alpha=1, c = 'lightGreen', edgecolor = 'black')

scatter.set_title("Dispersión del promedio por índice de crimen", size=16)

scatter.set_xlabel("Promedio", labelpad=20, weight='bold', size=12)

scatter.set_ylabel("Índice de crimen", labelpad=20, weight='bold', size=12)

scatter.set


plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
```

```python
barh = valor_por_ciudad.head(10).plot.barh(color='lightGreen', edgecolor = 'black')

barh.set_title("Distribución del promedio por ciudad", size=16)

barh.set_xlabel("Promedio", labelpad=20, weight='bold', size=12)

barh.set_ylabel("Ciudad", labelpad=20, weight='bold', size=12)

plt.show()
```
